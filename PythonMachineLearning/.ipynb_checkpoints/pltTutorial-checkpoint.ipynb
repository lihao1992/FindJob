```python
import numpy as np
from matplotlib import cm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

delta = 0.2
x = np.arange(-3, 3, delta)
y = np.arange(-10, 10, delta)
X, Y = np.meshgrid(x, y)
Z = (X + Y)**2

x=X.flatten()
y=Y.flatten()
z=Z.flatten()

fig = plt.figure()
plt.xlabel('x')
plt.ylabel('y')
ax = fig.gca(projection='3d')
ax.plot_trisurf(x, y, z, cmap=cm.jet, linewidth=0.01)
plt.show()


Z = (0.01*X + 10*Y)**2
x=X.flatten()
y=Y.flatten()
z=Z.flatten()

fig = plt.figure()
plt.xlabel('x')
plt.ylabel('y')
ax = fig.gca(projection='3d')
ax.plot_trisurf(x, y, z, cmap=cm.jet, linewidth=0.01)
plt.show()
```

```python
plt.contour(X, Y, Z,30)
plt.show()
```

```python
plt.contourf(X, Y, Z,3)
plt.show()
```
